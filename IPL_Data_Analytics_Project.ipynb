```python
import pandas as pd
#Load matches dataset
matches=pd.read_csv('matches.csv')
#Show first 5 rows
matches.head()
```

```python
matches.info()
```

```python
matches.describe()
```

```python
matches.columns
```

```python
matches['winner'].value_counts()
```

```python
team_wins=matches['winner'].value_counts().reset_index()
team_wins.columns=['team','wins']
team_wins
```

```python
team_wins=team_wins.sort_values(by='wins',ascending=False)
team_wins
```

```python
team_wins.to_csv('team_wins.csv',index=False)
```

```python
matches['toss_match_winner']=matches['toss_winner']==matches['winner']
```

```python
matches['toss_match_winner'].value_counts()
```

```python
matches['toss_match_winner'].value_counts(normalize=True)*100
```

```python
toss_analysis=matches['toss_match_winner'].value_counts().reset_index()
toss_analysis. columns = ['toss_result','count']
toss_analysis.to_csv('toss_analysis.csv',index=False)
```
